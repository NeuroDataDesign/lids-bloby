```python
from tifffile import imread, imsave
from scipy.ndimage.filters import gaussian_filter
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
import numpy as np

figsize(9, 9)
plt.set_cmap('gray')
```

```python
img = imread('../data/s3617/tifs/cell_detection_4.tiff')

plt.imshow(img[75, :, :])
plt.show()

g_img = gaussian_filter(img, sigma=(0.5, 1, 1))
imsave('../data/s3617/tifs/cd4_gauss.tiff', g_img.astype(np.uint16))
```
